# How to apply a transformation

We're going to apply a blur transformation to an image.

## Generate a random image

```python
import numpy as np
import matplotlib.pyplot as plt

random_img = np.random.random((100, 100, 3)) * 255
random_img = random_img.astype(np.uint8)


plt.imshow(random_img)
```

## Apply transformation

```python
from app import transform

bboxes = np.array([
    [10, 10, 50, 50],
    [20, 20, 80, 80],
    ]
)
transformed_img = transform.apply_transform(image=random_img, bboxes=bboxes, transformation=transform.blur_transformation)
plt.imshow(transformed_img)
```
